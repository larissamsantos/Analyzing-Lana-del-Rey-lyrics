# Wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lyrics = pd.read_csv('lyrics.csv')
```

```python
all_words = ' '.join([text for text in lyrics.Lyrics])
word_cloud = WordCloud(width= 800, height= 500,max_font_size = 110, collocations = False).generate(all_words)
plt.figure(figsize=(10,7))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

# NLTK

```python
import nltk

nltk.download("all")
```

```python
from nltk import tokenize
```

```python
token_space = tokenize.WhitespaceTokenizer()

tokens = token_space.tokenize(all_words)
frequency = nltk.FreqDist(tokens)
frequency_df = pd.DataFrame({"Word": list(frequency.keys()),
                                   "Frequency": list(frequency.values())})
```

```python
frequency_df.nlargest(columns = "Frequency", n = 10)
```

```python
import seaborn as sns

plt.figure(figsize=(12,8))
ax = sns.barplot(data = frequency_df.nlargest(columns= "Frequency", n = 10), x = "Word", y = "Frequency", color = 'gray')
ax.set(ylabel = "Count")
plt.show()
```

```python
stopwords = nltk.corpus.stopwords.words("english")
```

```python
processed_words = list()
for lyric in lyrics.Lyrics:
    new_lyrics = list()
    lyrics_words = token_space.tokenize(lyric)
    for word in lyrics_words:
        if word not in stopwords:
            new_lyrics.append(word)
    processed_words.append(' '.join(new_lyrics))
```
